```python
import string
terms = []
for i in '#'+string.ascii_lowercase:
    print(i, end=" ")
    page = requests.get("https://www.techopedia.com/it-terms/"+i)
    soup = BeautifulSoup(page.content, 'html.parser')
    for tag in soup.find_all('a', class_='text-body'):
        terms.append(tag.getText())
```

```python
len(terms)
```

```python
from difflib import SequenceMatcher

def similar(a, b):
    return SequenceMatcher(None, a, b).ratio()
```

```python
similar("Apple","Appel")
```

```python
with open('techopedia.txt','w') as f:
    for term in terms:
        f.write(term+"\n")
```
